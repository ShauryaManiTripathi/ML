```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
import seaborn
```

```python
def step_function(z):
    return 1 if z >= 0 else 0

class Perceptron:
    def __init__(self, learning_rate=0.1, n_iterations=100):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = np.zeros(2)
        self.bias = 0
        
    def predict(self, X):
        z = np.dot(X, self.weights) + self.bias
        return step_function(z)
    
    def train(self, X, y):
        n_samples = X.shape[0]
        errors = []
        
        for _ in range(self.n_iterations):
            error = 0
            print(f"\nIteration: {_}")
            for i in range(n_samples):
                print(f"sample: {i}")
                print(f"Current bias: {self.bias}")
                print(f"Current weights: {self.weights}")
                prediction = self.predict(X[i])
                update = self.learning_rate * (y[i] - prediction)
                self.weights += update * X[i]
                self.bias += update
                error += int(update != 0.0)

            errors.append(error)

            if error == 0:
                break
                
        return errors
    
# input data
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0, 0, 0, 1])

# training
perceptron = Perceptron(learning_rate=0.1, n_iterations=100)
errors = perceptron.train(X, y)
print("Errors:", errors)
# accuracy check
correct = 0
for i in range(len(X)):
    prediction = perceptron.predict(X[i])
    correct += (prediction == y[i])
accuracy = correct / len(X) * 100

print("\nend Weights:", perceptron.weights)
print("end bias:", perceptron.bias)
print(f"\n%age Accuracy: {accuracy}%")

print("\nTesting AND gate:")
print("X1 X2 | Output")
print("-" * 13)
for x1, x2 in X:
    output = perceptron.predict(np.array([x1, x2]))
    print(f" {x1}  {x2} |   {output}")
```
